```python
import numpy as np
import pandas as pd
```

```python
from numpy.random import randn, randint, uniform, sample
```

```python
df = pd.DataFrame({'value1':randn(1000),'value2':randn(1000),'value3':randn(1000),'value4':randn(1000)}, index = pd.date_range('2021-01-01', periods = 1000), columns=['value1','value2','value3','value4'])
```

```python
df['value1']=df['value1'].cumsum()
df['value2']=df['value2'].cumsum()
df['value3']=df['value3'].cumsum()
df['value4']=df['value4'].cumsum()
df
```
